```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_roc_curve
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import mean_squared_error

df_train = pd.read_csv('train_final.csv')
df_test = pd.read_csv('test_final.csv')
```

```python
target = df_train["Y"]
id = df_test["Id"].values

df_train.drop(['Id','Y'],axis=1,inplace=True)
df_test.drop(['Id'],axis=1,inplace=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_train, target, test_size=0.33, random_state=1)
```

```python
from catboost import CatBoostClassifier

#para = { "iterations":713, "learning_rate":0.019,  "verbose":0, "max_depth":6, "scale_pos_weight":[0.8, 0.2], 'loss_function': 'Logloss', 'l2_leaf_reg':3 }
para = { "iterations":713, "learning_rate":0.019,  "verbose":0, "max_depth":6, "class_weights":[0.55,0.45] }


catboost = CatBoostClassifier(**para)
                              

catboost.fit(x_train, y_train)

y_proba = catboost.predict_proba(x_test)
roc_auc_score(y_test, y_proba[:,1])
```

```python
y_proba_train = catboost.predict_proba(x_train)
roc_auc_score(y_train, y_proba_train[:,1])
```

```python
from sklearn.model_selection import RandomizedSearchCV

ct = CatBoostClassifier(class_weights=[0.55, 0.45], l2_leaf_reg=1, verbose=0)

parameters =   {'depth' : [ 2, 3, 4, 5,6,7],
                 'learning_rate' : [0.019,0.042, 0.03,0.034, 0.045, 0.04, 0.036, 0.038, 0.032],
                'iterations'    : [ 850, 1000, 950, 970, 1200, 1100, 1300, 1250, 1150]
               }

randm_src = RandomizedSearchCV(estimator=ct, param_distributions = parameters,
                               cv = 10, n_iter = 20, n_jobs=-1)

randm_src.fit(x_train, y_train)

#returns the estimator with the best performance
print(randm_src.best_estimator_)
    
#returns the best score
print(randm_src.best_score_)

#returns the best parameters
print(randm_src.best_params_)
```
