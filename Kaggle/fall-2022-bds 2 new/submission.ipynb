```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

```python
# Read in the training data and test data
train = pd.read_csv('train_final.csv')
test = pd.read_csv('test_final.csv')
```

```python
id = test["Id"].values
```

```python
train
```

```python
X_train = train.drop(['Id','Y'], axis=1)
X_train
```

```python
y_train = train['Y']
y_train
```

```python
# Fit the model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
X_test = test.drop(['Id'], axis=1)
X_test
```

```python
y_pred = model.predict(X_test)
y_pred
y_pred_proba = model.predict_proba(X_test)
```

```python
submission = pd.DataFrame({'Id':id, 'Y':y_pred_proba[:,1]})
submission["Id"] = submission["Id"].astype("Int32")
submission
```

```python
submission.to_csv('submission.csv', index=False)
```

```python
from sklearn.metrics import roc_auc_score
y_proba = model.predict_proba(X_test)
roc_auc_score(y_pred, y_proba[:,1])
```

```python
model_1 = RandomForestClassifier(n_estimators=100, max_depth=15, random_state=0)
model_1.fit(X_train, y_train)
```

```python
y_pred_1 = model_1.predict(X_test)
y_pred_1
y_pred_proba_1 = model_1.predict_proba(X_test)
submission = pd.DataFrame({'Id':id, 'Y':y_pred_proba_1[:,1]})
submission["Id"] = submission["Id"].astype("Int32")
submission
```
