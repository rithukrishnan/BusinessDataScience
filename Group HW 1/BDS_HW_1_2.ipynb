```python
import numpy as np
import math

mu = 0
sigma = 5
n = 25000

samples = np.random.normal(mu, sigma,n)

# print(samples)
```

```python
length = len(samples)
```

```python
total = sum(samples)
```

```python
mean = total / n
# print(mean)
```

```python
summation = 0
for i in samples:
    summation = summation + ((i - mean)**2)
    
# print(summation)

div_res = (summation / (n-1))

# print(div_res)

sd = math.sqrt(div_res)

```

### Results

```python
print("Mean : ",mean)
print("Standard Deviation: ",sd)
```
