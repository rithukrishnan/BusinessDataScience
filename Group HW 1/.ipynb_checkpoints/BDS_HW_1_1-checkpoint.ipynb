## Creating 1000 samples from a Gaussian distribution with mean -10 and standard deviation 5.


```python
import numpy as np

mu1 = -10
sd1 = 5
#Creating a random sample of 1000 using normal (Gaussian) distribution using numpy inbuilt random.normal method specifying the mean, standard deviation and the quantity
sample1 = np.random.normal(mu1, sd1, 1000)
```

```python
#Viewing the sample array created by the above method
sample1
```

```python
#Validating if 1000 samples are created for sample1
len(sample1)
```

```python
#Checking if the mean of the created gaussian distribution is approximately -10
sample1.mean()
```

```python
#Checking if the standard deviation of the created gaussian distribution is approximately 5
np.std(sample1)
```

## Creating 1000 samples from a Gaussian distribution with mean 10 and standard deviation 5.

```python
mu2 = 10
sd2 = 5
#Creating a random sample of 1000 using normal (Gaussian) distribution using numpy inbuilt random.normal method specifying the mean, standard deviation and the quantity
sample2 = np.random.normal(mu2, sd2, 1000)
```

```python
#Viewing the sample array created by the above method
sample2
```

```python
#Validating if 1000 samples are created for sample2
len(sample2)
```

```python
#Checking if the mean of the created gaussian distribution is approximately 10
sample2.mean()
```

```python
#Checking if the standard deviation of the created gaussian distribution is approximately 5
np.std(sample2)
```

```python
#Plotting histrograms for sample1 with matplotlib functions

import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(sample1, 30, density=True)
plt.plot(bins, 1/(sd1 * np.sqrt(2 * np.pi)) *
                np.exp( - (bins - mu1)**2 / (2 * sd1**2) ),
          linewidth=2, color='r')
plt.show()
```

```python
#Plotting histrograms for sample2 with matplotlib functions
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(sample2, 30, density=True)
plt.plot(bins, 1/(sd2 * np.sqrt(2 * np.pi)) *
                np.exp( - (bins - mu2)**2 / (2 * sd2**2) ),
          linewidth=2, color='r')
plt.show()
```

## Sum of these Gaussians by adding the two sets of 1000 points, point by point and plot the histogram of the resulting 1000 points. What do you observe?

```python
#Creating a new sample by adding the two sets of Gaussians with 1000 points, point by point
sample3 = sample1 + sample2
sample3
```

```python
#Validating if 1000 samples are created for sample3
len(sample3)
```

```python
#Checking if the mean of the new gaussian distribution is approximately 0 as we have added 2 gaussians, one with mean -10 and another with mean 10
mu3 = sample3.mean()
mu3
```

```python
#We have no presumed value of the standard deviation after adding both the gaussians point by point. We can compute the standard deviation with inbuilt numpy function
sd3 = np.std(sample3)
sd3
```

```python
#Calculating Varience
var3 = np.var(sample3)
var3
```

```python
#Plotting histrograms for sample3 with matplotlib functions
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(sample3, 30, density=True)
plt.plot(bins, 1/(sd3 * np.sqrt(2 * np.pi)) *
                np.exp( - (bins - mu3)**2 / (2 * sd3**2) ),
          linewidth=2, color='r')
plt.show()
```
