```python
#1. Create 1000 samples from a Gaussian distribution with mean -10 and standard deviation 5. Create another 1000 samples from another independent Gaussian with mean 10 and standard deviation 5.
#(a) Take the sum of these Gaussians by adding the two sets of 1000 points, point by point, and plot the histogram of the resulting 1000 points. What do you observe? Deliverables: three histograms (two for each Gaussian, one for the sum), written response, code
#(b) Estimate the mean and the variance of the sum.
#Deliverables: written response, code
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mu = -10
sigma = 5
s1 = np.random.normal(mu,sigma,1000)
plt.hist(s1, bins=25, density=True, alpha=0.6, color='b')
plt.show()
```

```python
mu = 10
sigma = 5
s2 = np.random.normal(mu,sigma,1000)
plt.hist(s2, bins=25, density=True, alpha=0.6, color='g')
plt.show()
```

# mu = 10
sigma = 5
s2 = np.random.normal(mu,sigma,1000)

```python
sum = np.add(s1,s2)
print(sum)
plt.hist(sum, bins=25, density=True, alpha=0.6, color='r')
plt.show()
```

```python
m = np.mean(sum)
print(m)
v = np.var(sum)
print(v)
```

```python
#Estimate the mean and standard deviation from 1 dimensional data: generate 25,000 samples from a Gaussian distribution with mean 0 and standard deviation 5. Then estimate the mean and standard deviation of this gaussian using elementary numpy commands, i.e., addition, multiplication, division (do not use a command that takes data and returns the mean or standard deviation).
#Deliverables: mean, standard deviation, code
```

```python
s3 = np.random.normal(0,5,25000)
#mean
m1 = np.sum(s3)/25000
print(m1)
```

```python
x = abs(s3 - s3.mean())**2
sum1 = np.sum(x)/(25000-1)
print(sum1)
```

```python
sd = np.sqrt(sum1)
```

```python
print("mean", m1)
print("standard deviation", sd)
```
